```python
#4.a
car_sales=pd.read_csv('car-sales.csv')
car_sales.info()
```

```python
car_sales['Price'].plot()
# Will not plot because 'Price' is not a numerical data type at the moment.
```

```python
car_sales['Odometer (KM)'].plot()
```

```python
#5.
car_sales['Price']=car_sales['Price'].str.replace('$', ' ')
car_sales['Price']=car_sales['Price'].str.replace(',',"")
```

```python
car_sales['Price']=car_sales['Price'].astype('int64')
```

```python
car_sales.info()
```

```python
#6
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

```python
#7
car_sales['Sale Date']=pd.date_range(start='2022-02-21', periods=len(car_sales))
car_sales.head()
```

```python
#8
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.title("Price vs Total Sales Line Plot")
plt.xlabel("Price")
plt.ylabel("Total Sales")
plt.show()
```

```python
#9
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title("Odometer vs Price Scatter Plot ")
plt.xlabel('Odometer (KM)')
plt.ylabel("Price")
plt.show()
```

```python
#10
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.title('Make vs Odometer Bar Plot')
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.show()
```

```python
#11
plt.hist(car_sales['Make'])
plt.title("Histogram of Car Make")
plt.xlabel('Make')
plt.show()
```

```python
#12
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2, figsize= (15,12))
ax1.plot(car_sales['Price'],car_sales['Total Sales'])
ax1.set_title("Price vs Total Sales Line Plot")
ax1.set_xlabel("Price")
ax1.set_ylabel("Total Sales")

ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
ax2.set_title("Odometer vs Price Scatter Plot ")
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel("Price")

ax3.bar(car_sales['Make'],car_sales['Odometer (KM)'])
ax3.set_title('Make vs Odometer Bar Plot')
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')

ax4.hist(car_sales['Make'])
ax4.set_title("Histogram of Car Make")
ax4.set_xlabel('Make')
```

```python
#13
import seaborn as sns
box_plot=sns.boxplot(x=car_sales['Colour'], y=car_sales['Total Sales'])
```

```python
#14
lm_plot=sns.lmplot(x='Total Sales',y='Odometer (KM)',data=car_sales,hue='Make',ci=False)
```

```python
#15
pair_plot=sns.pairplot(car_sales)
```

```python
#16
heat_map=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
axis=sns.heatmap(heat_map, annot=True)
#There is a negative correlation between price and Odometer. There is almost no correlation between Price and Total Sales
```
